```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
```

```python
import os

# Define the path to the folder containing all the folders with the csv files
#root_folder is path, where data is stored in local machine.
root_folder = '/Users/shashi/Downloads/SampleData27_01'

# Get the list of all the sub-folders in the root folder
sub_folders = [os.path.join(root_folder, folder) for folder in os.listdir(root_folder) if os.path.isdir(os.path.join(root_folder, folder))]

# Initialize a dictionary to store the sum of time for each step
time_sum = {}

# Initialize a dictionary to store the number of occurrences of each step
step_count = {}

# Loop through each sub-folder and calculate the average
df_average=pd.DataFrame()
for sub_folder in sub_folders:
    # Get the list of all csv files in this sub-folder
    csv_files = [os.path.join(sub_folder, file) for file in os.listdir(sub_folder) if file.endswith('TimeDataModule.csv')]
    
    #print(csv_files)
    # Loop through each csv file in this sub-folder
    all_lists = []
    for csv_file in csv_files:
        with open(csv_file, 'r') as f:
            # Read the data from the file 
            data = pd.read_csv(f,delimiter=',',usecols=range(2))
            #calculate the difference between the maxium and minimum time for each step
            average_steps=[]
            for s in data['step'].unique().tolist():
                time_stpes=data[data['step']==s]['time'].tolist()
                time_of_each_step=max(time_stpes)-min(time_stpes)
                average_steps.append(time_of_each_step)
                #print(average)
            
            
        #add the time of each step into the list  
        all_lists.extend(average_steps)
        
    # create a data frame from the lists
    df = pd.DataFrame(all_lists)

    # print the data frame
    df=df.T
    dictemp=[df]
    #pd.concat(df_average, df, axis=0)
    df_average=df_average.append(dictemp, ignore_index=True)
display(df_average)
   
            
        
         
```
